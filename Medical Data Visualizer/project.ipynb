# Medical Data Visualizer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Import dataframe

```python
df = pd.read_csv(r'../data/medical_examination.csv')
df.describe()
```

### Find if overweight

```python
BMI = df.weight / (df.height / 100) ** 2
df['overweight'] = (BMI > 25).astype(int)

df.overweight
```

### Normalize Data

```python
df.cholesterol = df.cholesterol.mask(df.cholesterol == 1, 0)
df.cholesterol = df.cholesterol.mask(df.cholesterol != 0, 1)

df.gluc = df.gluc.mask(df.gluc == 1, 0)
df.gluc = df.gluc.mask(df.gluc != 0, 1)

df[['cholesterol', 'gluc']]
```

### Catplot

```python
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

```python
cat_plot = sns.catplot(data = df_cat, kind = 'count',  x = 'variable', hue = 'value', col = 'cardio')
cat_plot.set_ylabels('total', fontsize=10)
```

### Clean data

```python
df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
                (df['height'] >= df['height'].quantile(0.025)) &
                (df['height'] <= df['height'].quantile(0.975)) &
                (df['weight'] >= df['weight'].quantile(0.025)) &
                (df['weight'] <= df['weight'].quantile(0.975))]

df_heat
```

### Correlation Matrix

```python
corr = df_heat.corr()
corr.round(1)

corr
```

```python
fig, heat_map = plt.subplots(figsize=(15, 10))

heat_map = sns.heatmap(corr, mask = np.triu(corr), annot = True, fmt = '.1f', square = True)
_ = heat_map.set_yticklabels(heat_map.get_yticklabels(), rotation = 0)
```

### Save plots

```python
cat_plot.savefig(r'catplot.png')
fig.savefig(r'heatmap.png')
```
